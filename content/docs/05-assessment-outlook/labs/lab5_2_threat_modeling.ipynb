````python
# ========== å¡«ç©º 3ï¼šé£é™©è¯„ä¼°çŸ©é˜µ ==========
# 
# ğŸ¯ ä»»åŠ¡ï¼šæ„å»ºé£é™©çŸ©é˜µï¼Œå¯¹æ‰€æœ‰å¨èƒè¿›è¡Œé‡åŒ–æ’åº
# 
# ğŸ’¡ æç¤ºï¼š
#   - é£é™©å€¼ = å¯èƒ½æ€§ Ã— å½±å“
#   - å°†å¨èƒæŒ‰é£é™©å€¼ä»é«˜åˆ°ä½æ’åº
#   - é£é™©å€¼ >= 12 ä¸ºé«˜é£é™©ï¼Œ>= 6 ä¸ºä¸­é£é™©ï¼Œå…¶ä½™ä¸ºä½é£é™©
# 
# è¯·å°† ___________ æ›¿æ¢ä¸ºæ­£ç¡®çš„ä»£ç 

def calculate_risk_matrix(stride_threats):
    """
    è®¡ç®—é£é™©çŸ©é˜µ
    
    å‚æ•°:
        stride_threats: STRIDE å¨èƒå­—å…¸
    
    è¿”å›:
        list: æŒ‰é£é™©å€¼æ’åºçš„å¨èƒåˆ—è¡¨
    """
    risk_list = []
    
    for category, threats in stride_threats.items():
        for threat in threats:
            risk_score = ___________
            # æœŸæœ›ï¼šè®¡ç®—é£é™©å€¼
            # æç¤ºï¼šthreat["likelihood"] * threat["impact"]
            
            if risk_score >= 12:
                risk_level = "é«˜"
            elif risk_score >= 6:
                risk_level = "ä¸­"
            else:
                risk_level = "ä½"
            
            risk_list.append({
                "id": threat["id"],
                "description": threat["description"],
                "category": category.split(" ")[0:3],
                "likelihood": threat["likelihood"],
                "impact": threat["impact"],
                "risk_score": risk_score,
                "risk_level": risk_level,
                "target_assets": threat["target_assets"]
            })
    
    # æŒ‰é£é™©å€¼é™åºæ’åˆ—
    risk_list.sort(key=lambda x: x["risk_score"], reverse=True)
    return risk_list

risk_matrix = calculate_risk_matrix(stride_threats)

# æ˜¾ç¤ºé£é™©çŸ©é˜µ
print("=" * 60)
print("ğŸ“Š é£é™©è¯„ä¼°çŸ©é˜µï¼ˆæŒ‰é£é™©å€¼æ’åºï¼‰")
print("=" * 60)
print(f"  {'ID':<4} {'å¨èƒæè¿°':<30} {'å¯èƒ½æ€§':>4} {'å½±å“':>4} {'é£é™©å€¼':>5} {'ç­‰çº§':<4}")
print(f"  {'â€”'*4} {'â€”'*30} {'â€”'*4} {'â€”'*4} {'â€”'*5} {'â€”'*4}")

for r in risk_matrix:
    icon = {"é«˜": "ğŸ”´", "ä¸­": "ğŸŸ¡", "ä½": "ğŸŸ¢"}[r["risk_level"]]
    desc = r["description"][:28] + "â€¦" if len(r["description"]) > 28 else r["description"]
    print(f"  {r['id']:<4} {desc:<30} {r['likelihood']:>4} {r['impact']:>4} {r['risk_score']:>5} {icon}{r['risk_level']}")

# ç»Ÿè®¡
high = sum(1 for r in risk_matrix if r["risk_level"] == "é«˜")
mid = sum(1 for r in risk_matrix if r["risk_level"] == "ä¸­")
low = sum(1 for r in risk_matrix if r["risk_level"] == "ä½")
print(f"\n  ç»Ÿè®¡ï¼šğŸ”´é«˜é£é™© {high} é¡¹ | ğŸŸ¡ä¸­é£é™© {mid} é¡¹ | ğŸŸ¢ä½é£é™© {low} é¡¹")
````

````python
# ========== å¡«ç©º 5ï¼šæ±‡æ€»ç”Ÿæˆå®Œæ•´æŠ¥å‘Š ==========
# 
# ğŸ¯ ä»»åŠ¡ï¼šå°†æ‰€æœ‰åˆ†æç»“æœæ±‡æ€»ä¸ºä¸€ä»½ç»“æ„åŒ–çš„å¨èƒå»ºæ¨¡æŠ¥å‘Š
# 
# ğŸ’¡ æç¤ºï¼š
#   - åŒ…å«æ‘˜è¦ã€èµ„äº§æ¸…å•ã€å¨èƒåˆ—è¡¨ã€é£é™©çŸ©é˜µã€ç¼“è§£å»ºè®®
#   - ç»™å‡ºæ€»ä½“é£é™©è¯„ä»·
#   - æå‡ºä¼˜å…ˆè¡ŒåŠ¨é¡¹
# 
# è¯·å°† ___________ æ›¿æ¢ä¸ºæ­£ç¡®çš„ä»£ç 

def generate_threat_report(app, assets, risk_matrix, mitigations):
    """ç”Ÿæˆå¨èƒå»ºæ¨¡æŠ¥å‘Š"""
    
    report_lines = []
    report_lines.append("=" * 60)
    report_lines.append("ğŸ“‹ AI åº”ç”¨å¨èƒå»ºæ¨¡æŠ¥å‘Š")
    report_lines.append("=" * 60)
    
    # 1. æ‘˜è¦
    report_lines.append(f"\nğŸ¯ è¯„ä¼°å¯¹è±¡ï¼š{app['name']}")
    report_lines.append(f"   æè¿°ï¼š{app['description']}")
    report_lines.append(f"   è¯„ä¼°æ—¶é—´ï¼š2025å¹´ï¼ˆæ¨¡æ‹Ÿï¼‰")
    
    # 2. èµ„äº§ç»Ÿè®¡
    high_assets = ___________
    # æœŸæœ›ï¼šç»Ÿè®¡é«˜æ•æ„Ÿåº¦èµ„äº§çš„æ•°é‡
    # æç¤ºï¼šsum(1 for a in assets if a["sensitivity"] == "é«˜")
    
    report_lines.append(f"\nğŸ“¦ èµ„äº§æ¦‚å†µï¼š")
    report_lines.append(f"   æ€»è®¡ {len(assets)} é¡¹èµ„äº§ï¼Œå…¶ä¸­ {high_assets} é¡¹é«˜æ•æ„Ÿåº¦")
    
    # 3. å¨èƒç»Ÿè®¡
    high_threats = sum(1 for r in risk_matrix if r["risk_level"] == "é«˜")
    mid_threats = sum(1 for r in risk_matrix if r["risk_level"] == "ä¸­")
    report_lines.append(f"\nğŸ” å¨èƒæ¦‚å†µï¼š")
    report_lines.append(f"   æ€»è®¡ {len(risk_matrix)} ä¸ªå¨èƒ")
    report_lines.append(f"   ğŸ”´ é«˜é£é™©ï¼š{high_threats} | ğŸŸ¡ ä¸­é£é™©ï¼š{mid_threats} | ğŸŸ¢ ä½é£é™©ï¼š{len(risk_matrix) - high_threats - mid_threats}")
    
    # 4. Top 3 ä¼˜å…ˆè¡ŒåŠ¨é¡¹
    report_lines.append(f"\nğŸš¨ ä¼˜å…ˆè¡ŒåŠ¨é¡¹ï¼ˆTop 3ï¼‰ï¼š")
    for i, threat in enumerate(risk_matrix[:3], 1):
        tid = threat["id"]
        m = mitigations.get(tid, {})
        report_lines.append(f"   {i}. [{tid}] {threat['description']}")
        report_lines.append(f"      é£é™©å€¼ï¼š{threat['risk_score']} | å»ºè®®ï¼š{m.get('strategy', 'éœ€è¦åˆ¶å®šç­–ç•¥')}")
    
    # 5. æ€»ä½“è¯„ä»·
    avg_risk = sum(r["risk_score"] for r in risk_matrix) / len(risk_matrix)
    if avg_risk >= 10:
        overall = "é«˜é£é™© â€” éœ€è¦ç«‹å³é‡‡å–è¡ŒåŠ¨"
    elif avg_risk >= 6:
        overall = "ä¸­ç­‰é£é™© â€” éœ€è¦åœ¨éƒ¨ç½²å‰å®Œæˆå…³é”®é¡¹ä¿®å¤"
    else:
        overall = "ä½é£é™© â€” åŸºæœ¬å®‰å…¨ï¼ŒæŒ‰è®¡åˆ’å®æ–½æ”¹è¿›å³å¯"
    
    report_lines.append(f"\nğŸ“Š æ€»ä½“é£é™©è¯„ä»·ï¼š")
    report_lines.append(f"   å¹³å‡é£é™©å€¼ï¼š{avg_risk:.1f}")
    report_lines.append(f"   è¯„ä»·ï¼š{overall}")
    
    report_lines.append(f"\n{'=' * 60}")
    report_lines.append("æŠ¥å‘Šç»“æŸ")
    report_lines.append("=" * 60)
    
    return "\n".join(report_lines)

# ç”Ÿæˆå¹¶æ‰“å°æŠ¥å‘Š
report_text = generate_threat_report(
    ai_application, assets, risk_matrix, mitigation_strategies
)
print(report_text)
````
